```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# ==== Condiciones iniciales ====
x0_cm = 1.0         # posición inicial x del centro de masa
y0_cm = 3.0         # posición inicial y del centro de masa
v0x = 5.0           # velocidad inicial horizontal del CM
v0y = 0.          # velocidad inicial vertical del CM
omega0 = 1.0        # velocidad angular inicial (rad/s)
theta0 = np.pi/4    # orientación inicial de la varilla (rad)
L = 1.0             # longitud de la varilla
g = 0             # gravedad

# ==== Tiempo de simulación ====
t_flight = 10  # estimación de tiempo de vuelo
t_vals = np.linspace(0, t_flight, 100)

# ==== Precalcular trayectorias ====
x1_sol, y1_sol = [], []
x2_sol, y2_sol = [], []
xcm_sol, ycm_sol = [], []

for t in t_vals:
    theta = theta0 + omega0 * t
    R = np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta),  np.cos(theta)]
    ])

    x_cm = x0_cm + v0x * t
    y_cm = y0_cm + v0y * t - 0.5 * g * t**2
    cm = np.array([[x_cm], [y_cm]])

    r1 = np.array([[L/2], [0.0]])     # extremo derecho en sistema de la varilla
    r2 = np.array([[-L/2], [0.0]])    # extremo izquierdo

    r1_abs = R @ r1 + cm
    r2_abs = R @ r2 + cm

    x1_sol.append(r1_abs[0, 0])
    y1_sol.append(r1_abs[1, 0])
    x2_sol.append(r2_abs[0, 0])
    y2_sol.append(r2_abs[1, 0])
    xcm_sol.append(x_cm)
    ycm_sol.append(y_cm)

# Convertir a arrays
x1_sol, y1_sol = np.array(x1_sol), np.array(y1_sol)
x2_sol, y2_sol = np.array(x2_sol), np.array(y2_sol)
xcm_sol, ycm_sol = np.array(xcm_sol), np.array(ycm_sol)

# ==== Visualización ====
fig, ax = plt.subplots(figsize=(6, 6))
traj1, = ax.plot([], [], 'k', label='Extremo 1')
traj2, = ax.plot([], [], 'g', label='Extremo 2')
traj_cm, = ax.plot([], [], 'ro', label='Centro de masa')

def init():
    x_min = min(x1_sol.min(), x2_sol.min(), xcm_sol.min()) - 0.5
    x_max = max(x1_sol.max(), x2_sol.max(), xcm_sol.max()) + 0.5
    y_min = min(y1_sol.min(), y2_sol.min(), ycm_sol.min()) - 0.5
    y_max = max(y1_sol.max(), y2_sol.max(), ycm_sol.max()) + 0.5

    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_xlabel('x(t)')
    ax.set_ylabel('y(t)')
    ax.set_title('Trayectorias de los extremos y el centro de masa')
    ax.legend()
    return traj1, traj2, traj_cm

def update(i):
    traj1.set_data(x1_sol[:i], y1_sol[:i])
    traj2.set_data(x2_sol[:i], y2_sol[:i])
    traj_cm.set_data(xcm_sol[:i], ycm_sol[:i])
    return traj1, traj2, traj_cm

ani = FuncAnimation(fig, update, frames=len(t_vals), init_func=init, blit=True, interval=30)
plt.close()
HTML(ani.to_jshtml())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# ==== Condiciones iniciales ====
x0_cm = 0         # posición inicial x del centro de masa
y0_cm = 9         # posición inicial y del centro de masa
v0x = 2.0           # velocidad inicial horizontal del CM
v0y = -2.0          # velocidad inicial vertical del CM
omega0 = -6.0        # velocidad angular inicial (rad/s)
theta0 = np.pi/4    # orientación inicial de la varilla (rad)
L = 2.0             # longitud de la varilla
g = 0          # gravedad

# ==== Tiempo de simulación ====
t_flight = 6
t_vals = np.linspace(0, t_flight, 100)

# ==== Precalcular posiciones ====
x1_sol, y1_sol = [], []
x2_sol, y2_sol = [], []
xcm_sol, ycm_sol = [], []

for t in t_vals:
    theta = theta0 + omega0 * t
    R = np.array([
        [np.cos(theta),  +np.sin(theta)],
        [-np.sin(theta),  np.cos(theta)] # Rotacion Horaria
    ])

    x_cm = x0_cm + v0x * t
    y_cm = y0_cm + v0y * t - 0.5 * g * t**2
    cm = np.array([[x_cm], [y_cm]])

    r1 = np.array([[L/2], [0.0]])
    r2 = np.array([[-L/2], [0.0]])

    r1_abs = R @ r1 + cm
    r2_abs = R @ r2 + cm

    x1_sol.append(r1_abs[0, 0])
    y1_sol.append(r1_abs[1, 0])
    x2_sol.append(r2_abs[0, 0])
    y2_sol.append(r2_abs[1, 0])
    xcm_sol.append(x_cm)
    ycm_sol.append(y_cm)

x1_sol, y1_sol = np.array(x1_sol), np.array(y1_sol)
x2_sol, y2_sol = np.array(x2_sol), np.array(y2_sol)
xcm_sol, ycm_sol = np.array(xcm_sol), np.array(ycm_sol)

# ==== Visualización solo de posición instantánea ====
fig, ax = plt.subplots(figsize=(6, 6))
line_segment, = ax.plot([], [], 'k-o', lw=3, label='Varilla')

def init():

    ax.set_xlim(0,10)
    ax.set_ylim(0,10)
    ax.set_aspect('equal')
    ax.set_xlabel('x')
    ax.set_ylabel('y')

    ax.legend()
    return line_segment,

def update(i):
    ax.set_title('Movimiento de la varilla (posición instantánea) {:.2f} s'.format(t_vals[i]))
    x_vals = [x2_sol[i], xcm_sol[i], x1_sol[i]]
    y_vals = [y2_sol[i], ycm_sol[i], y1_sol[i]]
    line_segment.set_data(x_vals, y_vals)
    return line_segment,

ani = FuncAnimation(fig, update, frames=len(t_vals), init_func=init, blit=True, interval=20)
plt.close()
HTML(ani.to_jshtml())
```

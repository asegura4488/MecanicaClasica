```python
# ============================================
#  Péndulo de Foucault 2D – versión corregida
#  (vectorizado, sin EDO)
# ============================================

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, HTMLWriter
from IPython.display import HTML

# ---------- PARÁMETROS ----------------------
g         = 9.81            # gravedad (m/s²)
L         = 10.0            # longitud del hilo (m)
A         = 2.0             # amplitud inicial (m)
Omega_sys = 2*np.pi/86164   # velocidad angular del sistema (rad/s)
Omega_sys = 1
lat_deg   = 30.0            # latitud (°)
Omega_sys *= np.cos(np.radians(lat_deg))

fps  = 5                  # fotogramas por segundo
dur  = 30                  # duración total (s)

# ---------- PRECÁLCULOS ---------------------
w0   = np.sqrt(g/L)                              # frecuencia natural
t    = np.linspace(0, dur, dur*fps + 1)          # tiempos
phiR = Omega_sys * t  # ángulo del plano

x = A * np.cos(w0 * t) * np.cos(phiR)            # x(t)
y = A * np.cos(w0 * t) * np.sin(phiR)            # y(t)

# ---------- FIGURA Y ANIMACIÓN --------------
fig, ax = plt.subplots(figsize=(5, 5))
ax.set_aspect('equal')
lim = A * 1.1
ax.set_xlim(-lim, lim); ax.set_ylim(-lim, lim)
ax.set_xlabel('x [m]'); ax.set_ylabel('y [m]')
ax.set_title('Péndulo de Foucault 2D')

line, = ax.plot([], [], lw=1, color='royalblue')
bob,  = ax.plot([], [], 'ro', ms=6)

def init():
    line.set_data([], [])
    bob.set_data([], [])
    return line, bob

def update(i):
    line.set_data(x[:i+1], y[:i+1])        # trayectoria completa hasta t_i
    bob.set_data([x[i]], [y[i]])           # <-- valores envueltos en listas
    return line, bob

ani = FuncAnimation(fig, update, frames=len(t),
                    init_func=init, blit=True, interval=1000/fps)
plt.close()

# ---------- SALIDA COMO VÍDEO HTML5 ---------
HTMLWriter.embed_limit = 40  # MB
HTML(ani.to_html5_video())
```

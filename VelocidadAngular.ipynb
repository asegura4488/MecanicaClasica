```python
import sympy as sp

# Variables simb├│licas
phi, theta, psi = sp.symbols('phi theta psi', real=True)
phi_dot, theta_dot, psi_dot = sp.symbols('phi_dot theta_dot psi_dot', real=True)

# Matrices de rotaci├│n
Rz_phi = sp.Matrix([
    [sp.cos(phi), -sp.sin(phi), 0],
    [sp.sin(phi),  sp.cos(phi), 0],
    [0, 0, 1]
])

Ry_theta = sp.Matrix([
    [1, 0, 0],
    [0, sp.cos(theta), -sp.sin(theta)],
    [0, sp.sin(theta), sp.cos(theta)]
])

Rz_psi = sp.Matrix([
    [sp.cos(psi), -sp.sin(psi), 0],
    [sp.sin(psi),  sp.cos(psi), 0],
    [0, 0, 1]
])

# Matriz total de rotaci├│n
R = Rz_phi * Ry_theta * Rz_psi

# Derivada temporal de R
R_dot = R.diff(phi) * phi_dot + R.diff(theta) * theta_dot + R.diff(psi) * psi_dot

# Calcular ¤Ѕ╠ѓ = Rрхђ ┬и R╠Є
omega_hat = R.T * R_dot

# Extraer componentes del vector de velocidad angular en la base del cuerpo
omega_1 = sp.simplify(omega_hat[2,1])
omega_2 = sp.simplify(omega_hat[0,2])
omega_3 = sp.simplify(omega_hat[1,0])

omega_vector = sp.Matrix([omega_1, omega_2, omega_3])

omega_hat = sp.simplify(omega_hat)

# Mostrar el vector de forma bonita
sp.init_printing(use_unicode=True)
display(sp.Eq(sp.Symbol('vec{¤Ѕ}'),
              sp.Matrix([
                  [omega_1],
                  [omega_2],
                  [omega_3]
              ])))
```

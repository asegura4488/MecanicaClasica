```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = (0,1.5)
x = np.linspace(-5,5,100)
y = x**2/(4*p[1])
plt.plot(x,y,color='k')
plt.plot(p[0],p[1],'bo')
plt.axhline(y=-p[1],color='r',linestyle='--')
plt.grid()
plt.show()
```

```python
def Elipse(a = 5, e=0.6):

  b = a * np.sqrt( 1 - e**2)
  c = a*e

  t = np.linspace(0, 2*np.pi, 100)

  x = a * np.cos(t)
  y = b * np.sin(t)

  foco1=(+c,0)
  foco2=(-c,0)

  plt.plot(x,y)
  plt.plot(foco1[0],foco1[1], 'ro')
  plt.plot(foco2[0],foco2[1], 'ro')
  plt.axis('equal')
  plt.grid()
  plt.show()
```

```python
Elipse(a=1,e=0.6)
```

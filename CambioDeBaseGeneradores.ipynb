```python
import sympy as sp

# Variables simb├│licas
phi, theta, psi = sp.symbols('phi theta psi', real=True)
phi_dot, theta_dot, psi_dot = sp.symbols('phi_dot theta_dot psi_dot', real=True)

# Matrices de rotaci├│n
Rz_phi = sp.Matrix([
    [sp.cos(phi), -sp.sin(phi), 0],
    [sp.sin(phi),  sp.cos(phi), 0],
    [0, 0, 1]
])

Ry_theta = sp.Matrix([
    [1, 0, 0],
    [0, sp.cos(theta),-sp.sin(theta)],
    [0, sp.sin(theta), sp.cos(theta)]
])

Rz_psi = sp.Matrix([
    [sp.cos(psi), -sp.sin(psi), 0],
    [sp.sin(psi),  sp.cos(psi), 0],
    [0, 0, 1]
])
```

```python
L1 = sp.Matrix([
    [0, 0, 0],
    [0, 0,-1],
    [0, 1, 0]
])
L2 = sp.Matrix([
    [0, 0, 1],
    [0, 0, 0],
    [-1, 0, 0]
])
L3 = sp.Matrix([
    [0, -1,0],
    [1, 0, 0],
    [0, 0, 0]
])
sp.init_printing(use_unicode=True)
L1,L2,L3
```

```python
A1 = Rz_psi.T * Ry_theta.T * Rz_phi.T * L3 * Rz_phi * Ry_theta * Rz_psi
A1 = sp.simplify(A1)
A1
```

```python
A2 = Rz_psi.T * L1 * Rz_psi
A2 = sp.simplify(A2)
A2
```

```python
A3 = Rz_phi.T * L3 * Rz_phi
A3 = sp.simplify(A3)
A3
```

```python
omegahat = phi_dot * A1 + theta_dot * A2 + psi_dot * A3
#omegahat = sp.simplify(omegahat)
omegahat
```

```python
w = sp.Matrix([-omegahat[1,2], omegahat[0,2],-omegahat[0,1]])
w
```

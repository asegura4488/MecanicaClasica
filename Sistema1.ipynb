```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
M = 5.0
m = 2.0
l = 1.0
g = 9.81

def System1(y,t):

  x, xdot, theta, thetadot = y

  I = np.array([[M+m, m*l*np.cos(theta)],
                [m*l*np.cos(theta), m*l**2]])

  F = np.array([m*l*thetadot**2*np.sin(theta), -m*g*l*np.sin(theta)])

  xddot, thetaddot = np.linalg.solve(I,F)

  return [xdot, xddot, thetadot, thetaddot]

# Poniendo friccion
def System2(y,t):


  x, xdot, theta, thetadot = y
  gamma = 1.0

  I = np.array([[M+m, m*l*np.cos(theta)],
                [m*l*np.cos(theta), m*l**2]])

  F = np.array([m*l*thetadot**2*np.sin(theta) - gamma*(2*xdot + l*thetadot*np.cos(theta)),
                -m*g*l*np.sin(theta) - gamma*(l*thetadot*np.cos(theta)+l**2*thetadot)])

  xddot, thetaddot = np.linalg.solve(I,F)

  return [xdot, xddot, thetadot, thetaddot]


# Tiempo de integración
t = np.linspace(0, 10, 100)
y0 = [0.,0.,0.5*np.pi,0.]

# Aca cambiamos al caso ideal y con friccion
sol = odeint(System2, y0, t)
x = sol[:,0]
theta = sol[:,2]
vx = sol[:,1]
vtheta = sol[:,3]

x_pen = x + l * np.sin(theta)
y_pen = -l * np.cos(theta)

vx_pen = vx + l * np.cos(theta) * vtheta
vy_pen = l * np.sin(theta) * vtheta

fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111)

def init():
  ax.set_xlim(-1,1)
  ax.set_ylim(-1,1)
  ax.set_xlabel('x(t)')
  ax.set_ylabel('y(t)')

def Update(i):
  ax.clear()
  ax.set_title('Trayectoria {:.2f}'.format(t[i]))
  init()
  ax.scatter(x[i],[0],color='r')
  ax.scatter(x_pen[i],y_pen[i],color='b')
  ax.plot([x[i],x_pen[i]],[0,y_pen[i]],'-',color='k')

  # Poner la velocidad
  ax.quiver(x[i],[0], vx[i], 0, angles='xy', scale_units='xy', scale=1, color='r')
  ax.quiver(x_pen[i],y_pen[i], vx_pen[i], vy_pen[i], angles='xy', scale_units='xy', scale=1, color='b')
  return []

# Crear animación
ani = FuncAnimation(fig, Update, frames=len(t), init_func=init)

# Mostrar en Colab
HTML(ani.to_jshtml())
```

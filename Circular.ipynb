```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = np.array([1,0])
```

```python
R = 1
omega1 = 1
omega2 = 2.
t = 1
```

```python
r1 = np.array([R*np.cos(omega1*t),R*np.sin(omega1*t)])
r2 = np.array([R*np.cos(omega2*t),R*np.sin(omega2*t)])
```

```python
v1 = omega1*R*np.array([-np.sin(omega1*t),np.cos(omega1*t)])
v2 = omega2*R*np.array([-np.sin(omega2*t),np.cos(omega2*t)])

print(v1)
print(v2)
```

```python
a1 = -omega1**2*R*np.array([np.cos(omega1*t),np.sin(omega1*t)])
a2 = -omega2**2*R*np.array([np.cos(omega2*t),np.sin(omega2*t)])
```

```python
np.linalg.norm(a2)
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111)
ax.set_xlim(-2,2)
ax.set_ylim(-2,2)
ax.arrow(0,0,r1[0],r1[1],head_width=0.1,length_includes_head=True,color='k')
ax.arrow(r1[0],r1[1],v1[0],v1[1],head_width=0.1,length_includes_head=True,color='b')
ax.arrow(r1[0],r1[1],a1[0],a1[1],head_width=0.1,length_includes_head=True,color='r')


ax.arrow(0,0,r2[0],r2[1],head_width=0.1,length_includes_head=True,color='k')
ax.arrow(r2[0],r2[1],v2[0],v2[1],head_width=0.1,length_includes_head=True,color='b')
ax.arrow(r2[0],r2[1],a2[0],a2[1],head_width=0.1,length_includes_head=True,color='r')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 1
```

```python
z = np.linspace(-L-0.5,L+0.5,100)
```

```python
def GetB(z,L,a):

  Campo = ( L/2 -  z )/ np.sqrt(  (L/2 - z)**2 + a**2  ) + ( L/2 + z )/ np.sqrt(  (L/2 + z)**2 + a**2  )

  return  Campo
```

```python
B = GetB(z,1,0.5)
```

```python
A = np.array([0.01,0.05,0.1,0.5,0.8,1])
```

```python
for a in A:
  plt.plot(z,GetB(z,L,a),label=r'{}'.format(a))

plt.legend()
```
